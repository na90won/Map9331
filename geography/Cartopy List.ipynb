```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

ax = plt.axes(projection =ccrs.PlateCarree())  #PlateCarree
ax.coastlines()

plt.show()
```

```python

ax = plt.axes(projection =ccrs.PlateCarree())
ax.stock_img() ##new added layer
ax.coastlines()

plt.show()
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

plt.figure(figsize=(5.12985642927, 3))
ax = plt.axes(projection=ccrs.AlbersEqualArea())  ##AlbersEqualArea
ax.coastlines(resolution='110m')
ax.gridlines()
```

```python

plt.figure(figsize=(5.12985642927, 5))
ax = plt.axes(projection=ccrs.AlbersEqualArea())  ##AlbersEqualArea
ax.coastlines(resolution='110m')
ax.stock_img() #layered as above
ax.gridlines()
```

```python

plt.figure(figsize=(5, 5))
ax = plt.axes(projection=ccrs.AzimuthalEquidistant(central_latitude=90))
ax.coastlines(resolution='110m')
ax.gridlines()
```

```python
plt.figure(figsize=(4.28969332205, 3))
ax = plt.axes(projection=ccrs.LambertConformal())
ax.coastlines(resolution='110m')
#without gridlines
```
