```python
import folium
m = folium.Map(location = [37.886071, 127.738089],
               tiles = "Stamen Terrain",
               zoom_start = 12)
tooltip = "Here!" # tooltip character

folium.Marker([37.839742, 127.761954], popup = '<b>dbox</b>', # dbox's location 
                 tooltip=tooltip).add_to(m)

folium.Marker([37.859228, 127.736908], popup = '<b>Seven7even</b>', #seveneleven cvs
                tooltip=tooltip).add_to(m)
m
```

```python
import folium
m = folium.Map(location = [37.886071, 127.738089],
               tiles = "Stamen Terrain",
               zoom_start = 12)


folium.Marker([37.839742, 127.761954], 
              popup = 'dbox',
              icon = folium.Icon(icon='cloud')).add_to(m)

folium.Marker([37.859228, 127.736908], 
              popup = 'Seven7even',
              icon = folium.Icon(icon = 'info-sign', color = 'red')).add_to(m)
folium.Marker([37.930495, 127.723271],
              pop = "ShindongSchool",
              icon = folium.Icon(icon ='cloud')).add_to(m)  # one of elementry school

m
```

```python
import folium
m = folium.Map(location = [37.886071, 127.738089],
               tiles = "OpenStreetMap",
               zoom_start = 12)

folium.CircleMarker([37.877772, 127.755078],
              pop = "Myhome",
              color = 'blue',
              fill = True,
              fill_color = 'blue',
              radius = 50,
              ).add_to(m) # near my apartment 


folium.CircleMarker([37.930495, 127.723271],
              pop = "Gangbook",
              color = 'red',
              fill = True,
              fill_color = 'red',
              radius = 100,
              ).add_to(m) # near school my bf come from

m.add_child(folium.LatLngPopup()) # help users to find a location by interactively browsing the map.
m.add_child(folium.ClickForMarker(popup='point')) # Marking functionality by each click
m
```
